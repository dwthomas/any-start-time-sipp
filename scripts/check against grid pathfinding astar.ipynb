```python
m = read_map("../../hog2/maps/rooms/32room_004.map")
```

|

```python
grid = []
for i in range(m.occupancy.shape[1]):
    grid.append(list(m.occupancy[i]))
```

```python
g = Grid(matrix = (1 -m.occupancy).T)
```

```python
start = g.node(6, 6)
end = g.node(445, 446)
```

```python
finder = AStarFinder(diagonal_movement=DiagonalMovement.only_when_no_obstacle)
```

```python
path, runs = finder.find_path(start, end, g)
```
